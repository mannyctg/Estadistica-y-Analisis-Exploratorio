### Distribucion de probabilidad de Variables Aleatorias Discretas

* Funcion masa de probabilidad ( PMF )

Es una función que asigna una probabilidad a cada valor posible de la variable, de forma que la suma de estas probabilidades sea igual a 1 y cada probabilidad individual esté entre 0 y 1.

* Funcion de distribucion acumulada ( CDF )

F(x) = Suma de las probabilidades hasta el valor de x incluyendo la probabilidad en el valor de
x.

### Valor Esperado
Es el promedio ponderado de todos los resultados posibles de un experimento aleatorio, donde cada resultado se pesa con su respectiva probabilidad

### Varianza 
Medida que cuantifica la dispersión o la variabilidad de los valores de esa variable con respecto a su media o valor esperado (u). Su raiz cuadrada se conoce como Desviacion Estandar.

### Distribuciones Bivariadas
- Son un método estadístico que analiza la relación entre dos variables aleatorias.
- Proporcionan las probabilidades de que los resultados de ambas variables ocurran al mismo tiempo.
- Permiten estudiar la asociación, correlación o dependencia entre dos factores. 

### Distribuciones Multivariadas
- Describen la probabilidad de que un conjunto de variables aleatorias (2 o mas) tome valores específicos al mismo tiempo. 

### Distribucion Marginal 
Proviene de un conjunto de datos que tiene múltiples variables, como una tabla de doble entrada o una distribución de probabilidad conjunta. Por ejemplo: Si en una encuesta se pregunta por el color favorito y el género, la distribución marginal de los colores indicaría cuántas personas prefieren el azul, el rojo, etc., sin especificar si son hombres o mujeres. La distribución marginal del género indicaría cuántos hombres y cuántas mujeres fueron encuestados, sin importar su color favorito. 

### Distribucion condicional
Muestra la distribución de valores de una variable dado que se conoce el valor de otra variable. 

Pasos para aplicarlo:
1. Identifica la variable de interés y la condición: Define cuál es la variable cuya distribución quieres analizar y cuál es la condición (la otra variable o su valor específico) que se toma como dada.
2. Selecciona el subconjunto de datos: Filtra o enfócate en la parte de los datos donde la condición se cumple. Por ejemplo, si quieres la distribución del porcentaje de aciertos para estudiantes que estudiaron 41-60 minutos, te enfocas solo en ese grupo de estudiantes.
3. Calcula la distribución dentro de ese subconjunto: Determina las frecuencias o probabilidades de los diferentes valores de la variable de interés solo dentro de este grupo. Esto se hace generalmente dividiendo las frecuencias dentro de ese subconjunto entre el tamaño total del subconjunto.

Modelo de probabilidad discreta que calcula la probabilidad de obtener un número específico de "éxitos" en una serie de ensayos independientes, donde cada ensayo solo tiene dos posibles resultados (éxito o fracaso).
