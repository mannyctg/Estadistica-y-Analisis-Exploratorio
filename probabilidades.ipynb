### Distribucion Marginal 
Proviene de un conjunto de datos que tiene mltiples variables, como una tabla de doble entrada o una distribucin de probabilidad conjunta. Por ejemplo: Si en una encuesta se pregunta por el color favorito y el gnero, la distribucin marginal de los colores indicara cuntas personas prefieren el azul, el rojo, etc., sin especificar si son hombres o mujeres. La distribucin marginal del gnero indicara cuntos hombres y cuntas mujeres fueron encuestados, sin importar su color favorito. 

### Distribucion condicional
Muestra la distribucin de valores de una variable dado que se conoce el valor de otra variable. 

Pasos para aplicarlo:
1. Identifica la variable de inters y la condicin: Define cul es la variable cuya distribucin quieres analizar y cul es la condicin (la otra variable o su valor especfico) que se toma como dada.
2. Selecciona el subconjunto de datos: Filtra o enfcate en la parte de los datos donde la condicin se cumple. Por ejemplo, si quieres la distribucin del porcentaje de aciertos para estudiantes que estudiaron 41-60 minutos, te enfocas solo en ese grupo de estudiantes.
3. Calcula la distribucin dentro de ese subconjunto: Determina las frecuencias o probabilidades de los diferentes valores de la variable de inters solo dentro de este grupo. Esto se hace generalmente dividiendo las frecuencias dentro de ese subconjunto entre el tamao total del subconjunto.
