# Regresion Lineal

* El analisis de regresion es una tecnica estadistica para modelar relaciones entre variables.

#### Variables en el analisis de regresion 

- y: variable respuesta o dependiente
- x: variable independiente

* Ecuacion de la recta: Y =  Bo + B1X

* La idea es encontrar la recta que mejor se ajuste a los datos.

* Vamos a usar la tecnica de los minimos cuadrados

#### Objetivos principales:
1. Encontrar valores que expliquen una situacion
2. Predecir valores

Y = Bo + B1X + E

* E es el termino de error, valor observado menos valor el valor en la recta

* Modelo mas adecuado: Modelo de regresion simple: un modelo de regresion lineal predice el valor de la variable dependiente Y a partir del valor de la variable independiente X.



* Y es la variable objetivo, es lo que queremos predecir
* X es la variable que conocemos, la tenemos en nuestros datos y es la que nos ayudara a predecir
* θ ( teta ) son los parametros, que son las variables fijas que se obtienen a partir de entrenar el modelo
* El modelo nos arroja los parametros, estos parametros vamos a reemplazarlos en la ecuacion de la recta y el resulta es la predicion de Y o variable objetivo.

#### Supuestos de la regresión lineal

1. Linealidad: la relación entre X y Y es más o menos lineal.

2. Homocedasticidad: los residuos deben tener varianza constante.

3. Normalidad de errores (sobre todo si haces inferencia).

4. Independencia de observaciones.
