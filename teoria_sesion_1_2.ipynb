Proviene de un conjunto de datos que tiene mltiples variables, como una tabla de doble entrada o una distribucin de probabilidad conjunta. Por ejemplo: Si en una encuesta se pregunta por el color favorito y el gnero, la distribucin marginal de los colores indicara cuntas personas prefieren el azul, el rojo, etc., sin especificar si son hombres o mujeres. ### Distribucion condicional
Muestra la distribucin de valores de una variable dado que se conoce el valor de otra variable. 

Pasos para aplicarlo:
1. Identifica la variable de inters y la condicin: Define cul es la variable cuya distribucin quieres analizar y cul es la condicin (la otra variable o su valor especfico) que se toma como dada.
2. Selecciona el subconjunto de datos: Filtra o enfcate en la parte de los datos donde la condicin se cumple. Por ejemplo, si quieres la distribucin del porcentaje de aciertos para estudiantes que estudiaron 41-60 minutos, te enfocas solo en ese grupo de estudiantes.
3. Calcula la distribucin dentro de ese subconjunto: Determina las frecuencias o probabilidades de los diferentes valores de la variable de inters solo dentro de este grupo. Esto se hace generalmente dividiendo las frecuencias dentro de ese subconjunto entre el tamao total del subconjunto.

## 7. Estimadores

- Un estimador es un estad stico (funci on que depende solo de los datos Xi).
- El estimador se usa con el fin de aproximar el valor verdadero de un parametro.
- La eleccion del estimador apropiado depende del par ametro poblacional de interes.

Se dividen en dos grupos: 

* Estimacion puntual: consiste en obtener un unico numero, calculado a partir de las
observaciones muestrales. Es utilizado como estimacion del valor del parametro .

* Estimacion por intervalos: se obtienen dos puntos (un extremo inferior y un extremo
superior) que definen un intervalo sobre la recta real, el cual contendra con cierta
seguridad el valor del parametro .

Propiedades de los estimadores:

* Insesgado (no se corre del valor real en promedio).

* Consistente (con ms datos, el estimador se acerca al valor real).

* Eficiente (poca variabilidad, es decir entre dos estimadores insesgados, el ms eficiente es el que tiene menor varianza).

* Suficiente (usa toda la info disponible).

* BONUS: los estimadores obtenidos por el metodo de Mxima verosimilitud (MLE) suelen tener buenas propiedades (OJO, no siempre).
